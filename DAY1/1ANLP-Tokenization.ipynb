```python
# Tokenization
# Tokenization is the process of breaking text into smaller pieces (tokens) 
# ‚Äî such as words, subwords, or characters 
# ‚Äî which are then converted into numbers for models to understand.
# Tokens are the basic units of text that models process.
Example sentence

"I love learning AI."

Tokenization can happen at different levels:
| Level           | Example Tokens                      | Description                        |
| --------------- | ----------------------------------- | ---------------------------------- |
| Word-level      | ["I", "love", "learning", "AI"]     | Each word is a token               |
| Subword-level   | ["I", "love", "learn", "ing", "AI"] | Breaks words into meaningful parts |
| Character-level | ["I", " ", "l", "o", "v", "e", ...] | Each character is a token          |
```

```python
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
#stopwords.words('english')
stemmer = PorterStemmer()
docs = nltk.sent_tokenize(corpus)  # paragrapgh -> sentence
print(type(docs),len(docs))

## Apply stopwords
## ----------------
for var in range(len(docs)):
    words = nltk.word_tokenize(docs[var])
    words = [stemmer.stem(word) for word in words if word 
             not in set(stopwords.words('english'))]
    docs[var] = " ".join(words)    
```

```python
ngrams
---------
1. What is an n-gram?

An n-gram is a contiguous sequence of n items from a given text or speech.
These "items" are usually words or characters.

Unigram (1-gram): single word

Bigram (2-gram): sequence of 2 words

Trigram (3-gram): sequence of 3 words

And so on‚Ä¶

N-grams are widely used in Natural Language Processing (NLP) for tasks like text prediction, 
spelling correction, and machine translation.


Why use n-grams?
Predict next word:
If we know bigrams, "I love" ‚Üí next word could be "machine", "you", etc.

Text similarity:
Compare overlapping n-grams between documents.

Language modeling:
Estimate probability of word sequences in NLP.
```

```python
TF-IDF stands for:

Term Frequency ‚Äì Inverse Document Frequency

It‚Äôs a way to measure how important a word is in 
a document relative to the entire dataset.

üß© Step 1: Recall Bag of Words
BoW only counts how often each word appears.
| Sentence                  | I | love | AI | machine | learning |
| ------------------------- | - | ---- | -- | ------- | -------- |
| "I love AI"               | 1 | 1    | 1  | 0       | 0        |
| "I love machine learning" | 1 | 1    | 0  | 1       | 1        |

How TF-IDF works

Term Frequency (TF):
       Total¬†terms¬†in¬†document
TF = ------------------------------
      Number¬†of¬†times¬†term¬†appears¬†in¬†document


Inverse Document Frequency (IDF):
DF=log( Total¬†number¬†of¬†documents / Number¬†of¬†documents¬†containing¬†the¬†term) 
‚Üí Common words get lower weight, rare words get higher weight.
TF-IDF score = TF √ó IDF
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

texts = [
    "I love AI",
    "I love machine learning"
]

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(texts)

print("Vocabulary:", vectorizer.get_feature_names_out())
print("TF-IDF Matrix:\n", X.toarray())
```

```python
Interpret the above result
| Word                      | Meaning                                            |
| ------------------------- | -------------------------------------------------- |
| **ai**                    | Appears only in sentence 1 ‚Üí higher importance     |
| **learning**, **machine** | Appear only in sentence 2 ‚Üí higher importance      |
| **love**                  | Appears in both ‚Üí lower importance                 |
| **I**                     | Ignored (stopword removed automatically by TF-IDF) |
```

```python
Comparison: BoW vs TF-IDF
| Feature             | Bag of Words               | TF-IDF                                |
| ------------------- | -------------------------- | ------------------------------------- |
| Values              | Raw counts (1, 2, 3‚Ä¶)      | Weighted (0‚Äì1)                        |
| Common words        | Treated equally            | Down-weighted                         |
| Rare words          | No special importance      | Up-weighted                           |
| Captures importance | ‚ùå                          | ‚úÖ                                     |
| Example Use         | Simple text classification | Information retrieval, search engines |
```

```python
now we‚Äôre stepping into modern NLP representation ‚Äî
the concept that came after Bag of Words and TF-IDF.

Word2Vec
=========
Let‚Äôs build your understanding of Word2Vec (Word to Vector) step by step.

What is Word2Vec?
Word2Vec = ‚ÄúWord to Vector‚Äù
It converts each word into a dense numerical vector 
(a list of numbers) that represents its meaning.

Unlike Bag of Words or TF-IDF (which only count words),
Word2Vec captures the relationship between words ‚Äî their context and similarity.

Example intuition
--------------------
In normal text, words that appear in similar contexts have similar meanings:
| Sentence examples  | Contextual idea            |
| ------------------ | -------------------------- |
| ‚ÄúKing and Queen‚Äù   | related to royalty         |
| ‚ÄúMan and Woman‚Äù    | related to gender          |
| ‚ÄúParis and France‚Äù | related to country/capital |

Word2Vec learns these relationships automatically.

Example of what Word2Vec learns
After training, words become vectors like this:
| Word  | Example Vector (simplified) |
| ----- | --------------------------- |
| king  | [0.52, 0.67, ‚Ä¶, -0.11]      |
| queen | [0.50, 0.70, ‚Ä¶, -0.09]      |
| man   | [0.40, 0.60, ‚Ä¶, -0.10]      |
| woman | [0.38, 0.63, ‚Ä¶, -0.08]      |

How Word2Vec works
It uses a neural network with one hidden layer.
Two main training methods:
| Method                             | Meaning                                                     |
| ---------------------------------- | ----------------------------------------------------------- |
| **CBOW (Continuous Bag of Words)** | Predicts the **current word** from nearby **context words** |
| **Skip-gram**                      | Predicts **context words** given the **current word**       |

üß± Simplified idea
If the text is:
‚ÄúI love learning AI‚Äù
Then:
Skip-gram: learns relationships like
("I" ‚Üí "love"), ("love" ‚Üí "learning"), ("learning" ‚Üí "AI")
CBOW: predicts ‚Äúlove‚Äù from context ["I", "learning"]
```

```python
from gensim.models import Word2Vec

# Sample data: list of tokenized sentences
sentences = [
    ["I", "love", "AI"],
    ["AI", "is", "the", "future"],
    ["I", "love", "machine", "learning"],
    ["machine", "learning", "is", "fun"]
]

# Train Word2Vec model
model = Word2Vec(sentences, vector_size=10, window=2, min_count=1, sg=1)  # sg=1 means skip-gram

# See vector for a word
print("Vector for 'AI':\n", model.wv["AI"])

# Find most similar words
print("\nWords similar to 'AI':")
print(model.wv.most_similar("AI"))
```

```python
Meaning:
Each word is now represented as a 10-dimensional vector.
‚ÄúAI‚Äù is closest to "learning" and "machine" ‚Äî makes sense!
```

```python
What is a Transformer?

Introduced in the paper ‚ÄúAttention is All You Need‚Äù (Vaswani et al., 2017)

Designed to process sequences (like text) in parallel, 
unlike RNNs or LSTMs which are sequential

Uses self-attention to understand relationships between all words in a sentence

Key components:
================
Input Embedding ‚Äì Converts words into dense vectors (like Word2Vec or learned embeddings)
Positional Encoding ‚Äì Adds position info since Transformers don‚Äôt inherently know word order
Encoder & Decoder ‚Äì Main building blocks (for tasks like translation)

Attention Mechanism ‚Äì Determines which words to focus on when processing each word

üîπ Self-Attention Mechanism

Self-Attention allows the model to look at all words 
in a sentence to compute a representation for a word.

Example:

Sentence:

‚ÄúThe cat sat on the mat.‚Äù

If we‚Äôre processing ‚Äúsat‚Äù, self-attention helps the model focus 
on ‚Äúcat‚Äù (subject) to understand meaning.

Mathematically:
Attention(Q,K,V) =  softmax(QKT/ root of dk) V
Where:
Q=Query
K=Key
V=Value

üîπ Transformer Architecture (Simplified)

Input Tokens ‚Üí Embeddings + Positional Encoding
             ‚Üì
         Encoder Layer (N times)
             ‚Üì
         Decoder Layer (N times, for tasks like translation)
             ‚Üì
         Output Probabilities (next word or translation)

Encoder ‚Üí Processes input sentence into context-aware representations
Decoder ‚Üí Generates output using encoder output + previous outputs (for autoregressive tasks)

Encoder Layer Components
=========================
Multi-Head Self-Attention ‚Üí Capture different types of relationships
Feedforward Neural Network ‚Üí Non-linear transformations
Residual Connections + Layer Norm ‚Üí Stabilize training

Decoder Layer Components
==========================
Masked Multi-Head Attention ‚Üí Only attends to previous words during generation
Encoder-Decoder Attention ‚Üí Focuses on encoder output
Feedforward + Residual + Layer Norm
```

```python
# Tokenization with a Transformer tokenizer (Hugging Face)
# This is how modern LLMs (like BERT, GPT, etc.) tokenize.
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

text = "I love learning AI."
tokens = tokenizer.tokenize(text)
token_ids = tokenizer.encode(text)

print("Tokens:", tokens)
print("Token IDs:", token_ids)
```

```python
# --- Install dependencies if needed ---
# pip install torch transformers scikit-learn matplotlib

import torch
from transformers import AutoTokenizer, AutoModel
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Load pretrained tokenizer & model
tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")
model = AutoModel.from_pretrained("distilbert-base-uncased")

# Some words to compare
words = ["dog", "puppy", "cat", "tiger", "car", "truck", "apple", "banana", "computer", "keyboard"]

# Get embeddings for each word
with torch.no_grad():
    input_ids = tokenizer(words, return_tensors="pt", padding=True)["input_ids"]
    embeddings = model.embeddings.word_embeddings(input_ids)

# Average subword embeddings (if a word gets split into multiple tokens)
word_embeddings = []
for i, word in enumerate(words):
    mask = input_ids[i] != tokenizer.pad_token_id
    emb = embeddings[i][mask].mean(dim=0)
    word_embeddings.append(emb.numpy())

# Reduce 768D ‚Üí 2D for plotting
pca = PCA(n_components=2)
emb_2d = pca.fit_transform(word_embeddings)

# Plot
plt.figure(figsize=(8, 6))
plt.scatter(emb_2d[:, 0], emb_2d[:, 1])

for i, word in enumerate(words):
    plt.text(emb_2d[i, 0] + 0.02, emb_2d[i, 1] + 0.02, word, fontsize=9)

plt.title("2D Visualization of Token Embeddings (DistilBERT)")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.grid(True)
plt.show()

```

```python
NLP Workflow: From Tokenization to Transformer
Step 1: Tokenization
       ‚îÇ
       ‚ñº
Step 2: Text Processing
       ‚îú‚îÄ‚îÄ Bag of Words (BoW)
       ‚îî‚îÄ‚îÄ N-grams
       ‚îÇ
       ‚ñº
Step 3: Word Embeddings (Deep Learning)
       ‚îî‚îÄ‚îÄ Word2Vec / GloVe / FastText
       ‚îÇ
       ‚ñº
Step 4: Neural Network
       ‚îî‚îÄ‚îÄ Feedforward or Deep NN for classification/regression
       ‚îÇ
       ‚ñº
Step 5: Contextual Word Embeddings
       ‚îî‚îÄ‚îÄ Dense, context-aware vectors for each word
       ‚îÇ
       ‚ñº
Step 6: Transformer
       ‚îú‚îÄ‚îÄ Encoder (BERT, etc.)
       ‚îú‚îÄ‚îÄ Decoder (GPT, translation)
       ‚îî‚îÄ‚îÄ Self-Attention captures relationships between all words
```

```python
Step-by-Step Explanation
Step 1: Tokenization
=====================
Break text into words, subwords, or sentences

Example: "I love AI" ‚Üí ["I", "love", "AI"]

Step 2: Text Processing
=========================
Bag of Words (BoW) ‚Üí count words in the text

N-grams ‚Üí sequences of n words (["I love", "love AI"])

Converts text ‚Üí numerical features for traditional ML

Step 3: Word Embeddings
=========================
Dense vector representation of words

Example: Word2Vec

"AI" ‚Üí [0.34, -0.12, 0.58, ...]

Captures semantic meaning of words

Words in similar contexts have similar vectors

Step 4: Neural Network
=========================
Takes embeddings or BoW features as input
Learns patterns or relationships for tasks:
Classification, Regression, Sentiment Analysis

Example: Predicting if "I love AI" is positive sentiment

Step 5: Contextual Word Embeddings
======================================
Classic embeddings are static ‚Üí same vector for ‚Äúbank‚Äù in ‚Äúriver bank‚Äù or ‚Äúmoney bank‚Äù

Contextual embeddings (from transformers) change vector based on sentence

Step 6: Transformer
=====================
Uses self-attention to understand the context of all words in the sentence

Encoder ‚Üí builds context-aware representations

Decoder ‚Üí generates output for translation, summarization, text generation

Example: BERT, GPT, T5

üîπ Visualization of Context Flow

Text: "I love AI"

Tokenization
   ‚Üì
["I", "love", "AI"]

BoW / N-grams
   ‚Üì
[1,1,1] or ["I love", "love AI"]

Word2Vec Embeddings
   ‚Üì
[[0.1,0.2,0.3], [0.3,0.4,0.1], [0.5,0.2,0.6]]

Neural Network
   ‚Üì
Predict Sentiment ‚Üí Positive

Transformer
   ‚Üì
Contextualized embeddings:
"I" = [0.12,0.09,...], "love" = [0.45,0.34,...], "AI" = [0.78,0.55,...]
Self-Attention captures relationships ‚Üí better prediction/generation
```
